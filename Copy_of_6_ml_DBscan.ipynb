```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/Mall_Customers.csv")
```

```python
data.head()
```

```python
df=data.drop(["CustomerID"],axis=1)
```

```python
df['Gender']=df['Gender'].apply({'Male':1,'Female':0}.get)
```

```python
df.head()
```

```python
from sklearn.cluster import DBSCAN
```

```python
cluster=DBSCAN(min_samples=5).fit(df)
```

```python
DBSCAN_dataset=df.copy()
```

```python
DBSCAN_dataset.loc[:,"cluster"]=cluster.labels_
```

```python
DBSCAN_dataset.cluster.value_counts().to_frame()
```
